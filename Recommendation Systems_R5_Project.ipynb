# Take subset of dataset to make it less sparse/more dense

```python
ratings_counts=ratings_electronics['user_id'].value_counts()
ratings_counts
```

```python
ratings_reduced = ratings_electronics[~ratings_electronics['user_id'].isin(ratings_counts[ratings_counts < 50].index)]
ratings_reduced
```

# Split the data randomly into train and test dataset

```python
from sklearn.model_selection import train_test_split
train_data,test_data = train_test_split(ratings_reduced, test_size=0.3, random_state=100)
```

# Popularity Recommender model

```python
import Recommenders as Recommenders
```

```python
pm = Recommenders.popularity_recommender_py()
pm.create(train_data,'user_id','prod_id')
```

```python
users = ratings_reduced['user_id'].unique()
```

```python
user_id = users[20]
pm.recommend(user_id)
```

# Collaborative Filtering mode

```python
is_model = Recommenders.item_similarity_recommender_py()
is_model.create(train_data, 'user_id', 'prod_id')
```

# Use the personalized model to make some song recommendations

```python
#Print the songs for the user in training data
user_id = users[5]
user_items = is_model.get_user_items(user_id)
#
print("------------------------------------------------------------------------------------")
print("Training data products for the user userid: %s:" % user_id)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_item)

print("----------------------------------------------------------------------")
print("Recommendation process going on:")
print("----------------------------------------------------------------------")

#Recommend songs for the user using personalized model
is_model.recommend(user_id)
```
