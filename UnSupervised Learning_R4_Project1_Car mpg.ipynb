```python
from scipy.stats import zscore
cars_df= pd.DataFrame(cars1.iloc[:,1:8],columns=['cyl', 'disp', 'hp', 'wt', 'acc', 'yr', 'mpg'])
cars_z= cars_df.apply(zscore)
```

```python
import seaborn as sns
sns.pairplot(cars_z.iloc[:,1:8],diag_kind='kde')
```

From the above pair plot we can see that the, the range of clusters may be from 3. hence consider the range of 
clusters from 2 to 8

```python
from sklearn.cluster import KMeans

cluster_range = range(2,8)   
cluster_errors = []
cluster_sil_scores = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 8, random_state=1)
  clusters.fit(cars_z.iloc[:,1:8])
  labels = clusters.labels_                     
  centroids = clusters.cluster_centers_         
  cluster_errors.append( clusters.inertia_ )    
  


# combine the cluster_range and cluster_errors into a dataframe by combining them
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors} )
clusters_df[0:15]


```

```python
# Elbow plot
from matplotlib import pyplot as plt
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

from the above elbow plot we can say that the significant drop in the error is less from between 3 to 5. the optimal number of clusters would be between 3 to 5
