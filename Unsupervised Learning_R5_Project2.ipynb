```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read input file

vehicle = pd.read_csv('vehicle.csv')
vehicle
```

DATA PRE PROCESSING

```python
vehicle.info()
```

```python
vehicle.head(10)
```

By observing the data, we can see that there are few NAN values in the given data. so missing values have to be 
treated. Replace the missing values by it's median

```python
vehicle.fillna(vehicle.median(),inplace=True)
vehicle
```

Analysis for the correlation between the variables

```python
plt.subplots(figsize = (19,19))
sns.heatmap(vehicle.corr(),annot=True)
```

```python
Box plot to check the outliers
```

```python

vehicle.plot(kind='box', subplots=True, layout=(5,5) ,fontsize=8, figsize=(20,20))
plt.show()
```
